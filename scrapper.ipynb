<h1> Scrapping du site  Bien dans ville

<h2> importation des librairies

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

<h2>Initialisation des listes

```python
date = []
commentaire = []
etoile = []
auteur = []
education=[]
securite=[]
loisir=[]
viePratique=[]
environnement=[]
arrondissement=[]
```

<h2> Scrapping des 20 arrondissements de Paris

```python
#On boucle sur les 20 arrondissements : 

for j in range(1,21):
    #on compare les url en string et si c'est un chiffre il faut un 0 devant
    if j<10 :
        j="0"+str(j)
    #on garde en mémoire l'adresse de la page 1 car elle est différente selon si on rentre une valeur dans l'url pour la page commentaire
    firstPage = "https://www.bien-dans-ma-ville.fr/paris-75056/arrondissement-"+str(j)+"/avis.html"
    
    #on liste toutes les pages que l'on veut scrapper
    pages=[]
    for i in range(1,11):
        pages.append("https://www.bien-dans-ma-ville.fr/paris-75056/arrondissement-"+str(j)+"/avis.html?page="+str(i)+"#commentaires")
    #on initialise des paramètres pour stopper le scrapping des qu'il n'y a plus de commentaires.
    i=0
    test=0;
    
    #on débute le scrapping page par page : 
    for page in pages:
        print(page)
        #ouverture de la page à scrapper
        page_SC = urlopen(page)
        soup = BeautifulSoup(page_SC, 'html.parser')
        if (test==1 or (page_SC.url == firstPage and i>0)):
            break
        else:
            i+=1
            if(len(soup.select(".commentaire")))<15:
                test= 1;
                
            #on entre dans les blocs commentaires
            for bloc in soup.select(".commentaire"):

                #arrondissement
                arrondissement.append(j)

                #date
                for element in bloc.select(".date"):
                    date.append(element.text)

                #commentaire
                for element in bloc.select(".description"):
                    commentaire.append(element.text)
                    
                #note_totale
                for element in bloc.select(".note_total"):
                    etoile.append(element.text)

                #auteur
                for element in bloc.select(".auteur"):
                    auteur.append(element.text)

                #securite
                for element in bloc.select("span:nth-child(1)"):
                    securite.append(element.text)

                #education
                for element in bloc.select("span:nth-child(2)"):
                    education.append(element.text)

                #loisir
                for element in bloc.select("span:nth-child(3)"):
                    loisir.append(element.text)

                #environnement
                for element in bloc.select("span:nth-child(4)"):
                    environnement.append(element.text)

                #vie pratique
                for element in bloc.select("span:nth-child(3)"):
                    viePratique.append(element.text)
                    
```

<h2> Mise au format dataframe

```python
#création du dataframe à partir des listes
scrap = pd.DataFrame(list(zip(date,arrondissement,auteur,etoile,education,securite,loisir,viePratique,environnement,commentaire)), columns=["date","arrondissement","auteur","etoile",
                                                                               "education","securite","loisir","viePratique",
                                                                              "environnement","commentaire"])
```

```python
#sauvegarde du dataframe
scrap.to_csv("bien_dans_ma_ville.csv")
```
