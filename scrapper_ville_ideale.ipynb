<h1> Scrapping du site  Bien dans ville

<h2> importation des librairies

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

<h2>Initialisation des listes

```python
date = []
commentairePositif = []
commentaireNegatif = []
etoile = []
auteur = []
education=[]
securite=[]
loisir=[]
viePratique=[]
environnement=[]
arrondissement=[]
transport=[]
sante=[]
culture=[]
commerce=[]
```

<h2> Scrapping des 20 arrondissements de Paris

```python
Paris = ["paris-1er-arrondissement_75101","paris-2e-arrondissement_75102","paris-3e-arrondissement_75103","paris-4e-arrondissement_75104","paris-5e-arrondissement_75105","paris-6e-arrondissement_75106","paris-7e-arrondissement_75107","paris-8e-arrondissement_75108","paris-9e-arrondissement_75109","paris-10e-arrondissement_75110","paris-11e-arrondissement_75111","paris-12e-arrondissement_75112","paris-13e-arrondissement_75113","paris-14e-arrondissement_75114","paris-15e-arrondissement_75115","paris-16e-arrondissement_75116","paris-17e-arrondissement_75117","paris-18e-arrondissement_75118","paris-19e-arrondissement_75119","paris-20e-arrondissement_75120"]
```

```python
#On boucle sur les 20 arrondissements : 

for arrdt in Paris:
    #on garde en mémoire l'adresse de la page 1 car elle est différente selon si on rentre une valeur dans l'url pour la page commentaire
    firstPage = "https://www.ville-ideale.fr/"+ arrdt
    
    #on liste toutes les pages que l'on veut scrapper
    pages=[]
    pages.append(firstPage)
    for i in range(2,20):
        pages.append("https://www.ville-ideale.fr/"+arrdt+"?page="+str(i)+"#commentaires")

    
    #on débute le scrapping page par page : 
    for page in pages:
        print(page)
        #ouverture de la page à scrapper
        page_SC = urlopen(page)
        soup = BeautifulSoup(page_SC, 'html.parser')
        if (page_SC.url == "https://www.ville-ideale.fr/erreur_404.php#commentaires"):
            break
        else:               
            #on entre dans les blocs commentaires
            for bloc in soup.select(".comm"):

                #arrondissement
                arrondissement.append(arrdt)

                #date
                for element in bloc.select(".comm span"):
                    date.append(element.text)

                #commentairePositif
                for element in bloc.select(".comm table+ p"):
                    commentairePositif.append(element.text)
                    
                #commentaireNegatif
                for element in bloc.select(".comm p+ p"):
                    commentaireNegatif.append(element.text)
                    
                #note_totale
                for element in bloc.select(".moyenne"):
                    etoile.append(element.text)

                #auteur
                for element in bloc.select("br+ strong"):
                    auteur.append(element.text)

                #securite
                for element in bloc.select("#colleft td:nth-child(3)"):
                    securite.append(element.text)

                #education
                for element in bloc.select("#colleft td:nth-child(7)"):
                    education.append(element.text)

                #loisir
                for element in bloc.select("#colleft td:nth-child(5)"):
                    loisir.append(element.text)
                    
                #santé
                for element in bloc.select("#colleft td:nth-child(4)"):
                    sante.append(element.text)
                
                #transport
                for element in bloc.select("#colleft td:nth-child(2)"):
                    transport.append(element.text)

                #environnement
                for element in bloc.select("#colleft td:nth-child(1)"):
                    environnement.append(element.text)

                #qualiteDeVie
                for element in bloc.select("#colleft td:nth-child(9)"):
                    viePratique.append(element.text)
                
                #Culture
                for element in bloc.select("#colleft td:nth-child(6)"):
                    culture.append(element.text)
                 
                #commerce
                for element in bloc.select("#colleft td:nth-child(8)"):
                    commerce.append(element.text)
                print(commerce) 
```

<h2> Mise au format dataframe

```python
#création du dataframe à partir des listes
scrap = pd.DataFrame(list(zip(date,arrondissement,auteur,etoile,education,securite,loisir,sante,transport,viePratique,culture,commerce,environnement,commentairePositif,commentaireNegatif)), columns=["date","arrondissement","auteur","etoile",
                                                                               "education","securite","loisir","sante","transport","qualite de vie","culture","commerce",
                                                                              "environnement","commentairePositif","commentaireNegatif"])
```
